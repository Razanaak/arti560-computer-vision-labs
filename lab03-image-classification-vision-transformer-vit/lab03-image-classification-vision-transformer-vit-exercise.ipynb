```python
# Import Libraries
import tensorflow as tf
import keras_hub
import numpy as np
from PIL import Image
import requests

# Load ViTImageClassifierPreprocessor (vit_base_patch16_224_imagenet preset)
# NOTE: This line may trigger an ImportError on Windows due to tensorflow-text versioning.
preprocessor = keras_hub.models.ViTImageClassifierPreprocessor.from_preset(
    "vit_base_patch16_224_imagenet"
)

# Load ViTImageClassifier (vit_base_patch16_224_imagenet preset)
model = keras_hub.models.ViTImageClassifier.from_preset(
    "vit_base_patch16_224_imagenet"
)

# Load the images
labels_url = "https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt"
imagenet_labels = requests.get(labels_url).text.splitlines()
image_files = ["cat.jpg", "laptop.jpg", "mug.jpg", "pizza.jpg", "car.jpg"]

# Predict classes
for file in image_files:
    img = Image.open(file).convert("RGB")
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)
    
    # Using the required preprocessor
    preprocessed_img = preprocessor(img_array)
    predictions = model.predict(preprocessed_img, verbose=0)
    
    predicted_idx = np.argmax(predictions, axis=-1)[0]
    print(f"File: {file} | Predicted: {imagenet_labels[predicted_idx]}")
```

The code above follows the official assignment instructions. However, due to a known compatibility issue between tensorflow-text and Windows (Python 3.13), it may raise an ImportError. I have provided a functional manual implementation in the next cell that uses the exact same mathematical normalization required by the ViT model."

```python
# Import Libraries
import os
os.environ["KERAS_BACKEND"] = "tensorflow" # Force tensorflow backend
import keras_hub
import numpy as np
from PIL import Image
import requests
import tensorflow as tf

# Load ViTImageClassifier (vit_base_patch16_224_imagenet preset)
# We set preprocessor=None to avoid the tensorflow-text error on Windows
model = keras_hub.models.ViTImageClassifier.from_preset(
    "vit_base_patch16_224_imagenet",
    preprocessor=None 
)

# Load the images
labels_url = "https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt"
imagenet_labels = requests.get(labels_url).text.splitlines()
image_files = ["cat.jpg", "laptop.jpg", "mug.jpg", "pizza.jpg", "car.jpg"]

# Predict classes
for file in image_files:
    try:
        # 1. Load and resize to 224x224
        img = Image.open(file).convert("RGB").resize((224, 224))
        img_array = np.array(img).astype("float32")
        
        # 2. Manual Normalization (Mandatory for ViT to avoid "nematode" result)
        # ViT expects pixels scaled via the formula: (x / 127.5) - 1
        preprocessed_img = (img_array / 127.5) - 1.0
        
        # 3. Add batch dimension (1, 224, 224, 3)
        input_data = np.expand_dims(preprocessed_img, axis=0)
        
        # 4. Predict
        predictions = model.predict(input_data, verbose=0)
        predicted_idx = np.argmax(predictions, axis=-1)[0]
        
        print(f"File: {file:12} | Predicted: {imagenet_labels[predicted_idx]}")
    except Exception as e:
        print(f"Error processing {file}: {e}")
```
